```
reference_set = set()
billb["Name"] = billb["Name"].str.lower()
billb["Artists"] = billb["Artists"].str.lower()
spotify["str_artists"] = [', '.join(map(str, l)).lower() for l in spotify["artists"]]
```

```
spotify["str_song"] = spotify["name"].str.lower()
rows = spotify.shape[0]
print(rows)
spotify["good_song"] = [0 for i in range(rows)]
spotify["good_artist"] = [0 for i in range(rows)]
```

```
def strip_accents(s):
   return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')

df_stripsong = []
df_stripname = []
for i in spotify['str_song']:
  df_stripsong.append(strip_accents(i))

for j in spotify['str_artists']:
  df_stripname.append(strip_accents(j))

spotify['str_song'] = df_stripsong
spotify['str_artists'] = df_stripname
```

```
df_stripsong = []
df_stripname = []
for i in streamv2['Name']:
  df_stripsong.append(strip_accents(i))

for j in streamv2['combArt']:
  df_stripname.append(strip_accents(j))

streamv2['str_name'] = df_stripsong
streamv2['combArt'] = df_stripname
streamv2
```

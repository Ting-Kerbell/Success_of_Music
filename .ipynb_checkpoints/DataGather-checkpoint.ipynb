```python
spotify = pd.read_csv(dataset_path+"data.csv")
billb = pd.read_csv(dataset_path+"billboardHot100_1999-2019.csv")
spotify = spotify[spotify['year']>1999]
spotify = spotify[spotify['year']<2010]
streams = pd.read_csv(dataset_path+"spotifyWeeklyTop200Streams.csv")
riaa = pd.read_csv(dataset_path+"riaaSingleCerts_1999-2019.csv")
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="INSERT ID",
                                                           client_secret="INSERT ID"))
```

```python
df1_grouped = billb.groupby(['Name', 'Artists'])
billb['ID']="0000000"
# iterate over each group
missingNum = 0
BillbSetMult = set()
for group_name, df_group in df1_grouped:
  print(group_name)
  songName = group_name[0]
  artistName = group_name[1]
  queryStr = songName+" artist:"+artistName
  results = sp.search(q=queryStr, limit=1)
  n = len(results['tracks']['items'])
  if n!=0:
    #billb.loc[(billb['Name'] == songName) & (billb['Artists'] == artistName), "ID"] = results['tracks']['items'][0]['id']
    for idx, track in enumerate(results['tracks']['items']):
      BillbSetMult.add(track['id'])
  else:
    missingNum+=1

```

```python
spotify['is_success']=0
billbSet = billb["ID"].unique()
print(len(BillbSetMult))
spotify.loc[(spotify["id"].isin(BillbSetMult)), 'is_success'] = 1
```
